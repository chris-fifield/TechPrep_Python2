```python
import pandas as pd
import koalas as kl

# Read CSV file containing student grades
grades = pd.read_csv("Grades_Short.csv")

# Get an average of all grades for each student and add it to new column "Average".
grades["Average"] = round(grades.loc[:, "Assignment_1":"Final_Exam"].mean(axis=1), 1)

# Pass average to function getLetterGrade() which returns a letter rank for each row. Store value in new column "Letter_Grade".
grades["Letter_Grade"] = grades["Average"].apply(lambda a: kl.getLetterGrade(a))

grades.to_csv("Grades_Updated.csv", index=False) # Save all data with the new columns added.

grades

```
